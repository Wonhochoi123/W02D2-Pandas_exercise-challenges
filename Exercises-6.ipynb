### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.
- first table is in USD
- second table actually consists of 3 smaller tables. First 9 rows are for USD, next 8 rows are for GBP and final 7 are for EUR.

```python
fir_USD=sales1.rename(columns={'Book title':'Title', 'Number sold': 'Units sold', 'Sales price': 'List price', 'Royalty paid': 'Royalty'})
fir_USD=fir_USD.set_index(['Title'])
fir_USD=fir_USD.rename(index={'Pining For The Fisheries of Yore':'Pining for the Fisheries of Yore','The Bricklayer’s Bible':"The Bricklayer's Bible"})
fir=fir_USD['Units sold']*fir_USD['List price']


sec_USD=sales2[:9]
sec_USD=sec_USD.set_index(['Title'])
sec_USD=sec_USD[3:-1]
sec=sec_USD['Units sold']*sec_USD['List price']

USD=sec+fir


sec_GBP=sales2[9:17]
sec_GBP=sec_GBP.set_index(['Title'])
sec_GBP=sec_GBP[4:-1]
GBP=sec_GBP['Units sold']*sec_GBP['List price']

sec_EUR=sales2[17:]
sec_EUR=sec_EUR.set_index(['Title'])
sec_EUR=sec_EUR[4:-1]
EUR=sec_EUR['Units sold']*sec_EUR['List price']
```

```python
df=pd.DataFrame({'USD':pd.Series(USD),'GBP':pd.Series(GBP),'EUR':pd.Series(EUR)})
df
```
